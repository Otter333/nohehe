```python
import feedparser
from google.colab import auth
from google.auth import default
import gspread

# 解析 Yahoo 新聞 RSS
feed_url = "https://tw.news.yahoo.com/rss"
feed_data = feedparser.parse(feed_url)

# Google 認證
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

# 創建 Google 試算表
spreadsheet = gc.create("Yahoo_News_Feed")
sheet = spreadsheet.sheet1
sheet.append_row(["title", "summary", "link"])

# 寫入 RSS 資料
for entry in feed_data.entries:
    sheet.append_row([entry.title, entry.summary, entry.link])

print(f"資料已成功寫入試算表: {spreadsheet.url}")
```

```python
url = 'https://docs.google.com/spreadsheets/d/1XY16V1u6gdWtJi6enHtU55CI8XRrDCYlqAfxcP1L_P4'
workbook = gc.open_by_url(url)
sheet = workbook.get_worksheet(0)

sheet.clear()
sheet.append_row(['title', 'summary', 'link'])
for entry in feed_data.entries:
    sheet.append_row([entry.title, entry.summary, entry.link])
```
